```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
feature = ['Year','Make','Used/New','Price','ConsumerRating','ConsumerReviews','SellerType','SellerRating','SellerReviews','State','Mileage','ExteriorColor','FuelType']
selectFeature = cars[feature]
selectFeature
```

```python
selectFeature.info()
```

```python
temp = cars['FuelType']
temp = list(set(temp))
temp
```

```python
elec = ['Electric Fuel System','Electric','Plug-In Electric/Gas']
hybird = ['Flexible Fuel','Hybrid','Gasoline/Mild Electric Hybrid']
trad = ['Gasoline Fuel', 'Diesel Fuel','Gasoline','Diesel']

selectFeature['usedYear'] = 2022-selectFeature['Year']
selectFeature['Mileage*Year'] = selectFeature['Mileage'] * selectFeature['usedYear'] / 100
selectFeature.info()
```

```python
visual = selectFeature[['Mileage*Year','Price','FuelType','Year']]
print(min(visual['Mileage*Year']))
print(max(visual['Mileage*Year']))
indexNoPrice = visual[visual['Price'].str.contains('Not Priced')].index
print(indexNoPrice)
visual.drop(indexNoPrice,inplace = True)
visual['Price2'] = visual.Price.str.replace('$','').str.replace(',','').astype('int')
indexOutli = visual[visual['Mileage*Year']>15000].index
visual.drop(indexOutli,inplace=True)
# indexOutli = visual[visual['Price2']>50000].index
# visual.drop(indexOutli,inplace=True)
visual
```

```python
elecData = visual[visual.FuelType.isin(elec)]
hybirdData = visual[visual.FuelType.isin(hybird)]
tradData = visual[visual.FuelType.isin(trad)]

elecData.info()
```

```python
nSample = 50
elecData = elecData.sample(n=nSample)
hybirdData = hybirdData.sample(n=nSample)
tradData = tradData.sample(n=nSample)
```

```python
plt.scatter(elecData['Mileage*Year'],elecData['Price2'],c='red',s=2,label='Electric')
plt.scatter(hybirdData['Mileage*Year'],hybirdData['Price2'],c='green',s=2,label='Hybrid')
plt.scatter(tradData['Mileage*Year'],tradData['Price2'],c='blue',s=2,label='Oil')
plt.legend()
plt.yticks(np.arange(0,250000,40000))
plt.ylabel("Price")
plt.xlabel("Year*Mileage")
plt.title("Price-Year*Mileage")
plt.figure(figsize=(500,500))
# plt.savefig('temp.png')
plt.show()
```

```python
lineStyle = '--'
plt.scatter(elecData['Year'],elecData['Price2'],color='red',label='Electric')
plt.scatter(hybirdData['Year'],hybirdData['Price2'],color='green',label='Hybird')
plt.scatter(tradData['Year'],tradData['Price2'],color='blue',label='Oil')
plt.legend()
plt.xticks(np.arange(2000,2024,2))
plt.ylabel("Price")
plt.xlabel("Year")
plt.title("Price-Year")
plt.figure(figsize=(500,500))
plt.show()
```
